## Izdrukājamās informācijas formatēšana

Parasti formatēšanu veic skaitļiem ar peldošo komatu:

```python
pi = 3.14159 
print(pi)
```

Formatēšanai izmanto papildus metodi “format” , kas tiek piemērot formatējumam, kur figūriekavu pāri nozīmē vietas, kur ievietot datus, bet paši dati padoti kā parametri:

```python
pi=3.14159
print("{} - {}".format(pi,3))
```

Datu pozīcijas var mainīt, ievietojot numurus figūriekavās (parametros datu pozīcijas numurējas 0, 1, ...):

```python
pi=3.14159
print("{1} + {0}".format(pi,3))
```

Izmantojot papildus modifikatoru “f” , var noteikt ciparu skaitu aiz komata:

```python
pi=3.14159
print("{0:.2f}".format(pi))
```

Izmantojot papildus modifikatoru “e” , izvade tiek formatēta zinātniski:

```python
pi=3.14159
print("{0:.2e} {0:.4e}".format(pi,pi))
```

Teksta formatēšanā var izmantot, ka operators “*” nozīmē teksta atkārtošanu n reizes:

```python
print("wow"*5)
```

```python
#primitīvs aprēķina kods

a = 4
b = 7

# * - reiznat
# / - dalit
# - - atnemt
# + - saskaitit

c = a - b

print(c)
```

# Masīvs un saraksti

```python
# Saraksta izveidošana, kurš satur elementus, kuri pieder dažādiem datu tipiem  
saraksts = 1, 2, "Lalala", 'a','e' 
print(saraksts) 
```

```python
import array as array
masivs = array.array('i', [1, 2, 3])   
# piekļūt masīva elementiem 
for i in masivs: 
    print(i) 
```

# Loģiskie operatori

Dažādas loģiskās vērtības, piemēram, salīdzināšanas operācijas var apvienot, veidojot sarežģītākas loģiskās izteiksmes. Šim nolūkam der loģiskie operatori “not and or”.

```python
# pasaka, vai dotais skaitlis ir pozitīvs viencipara skaitlis:
a = int(input("Ievadi skaitli: "))
print(a >= 0 and a <= 9)
```

```python
# pasaka, vai dotais skaitlis nav pozitīvs viencipara skaitlis:
a = int(input("Ievadi skaitli: "))
print (a < 0 or a > 9)
```

```python
# pasaka, vai dotais skaitlis ir pozitīvs viencipara skaitlis
a = int(input("Ievadi skaitli: "))
print (not(a < 0 or a > 9))
```

# Cikla konstrukcijas

Cikls (looping) ir vadības konstrukcija, kas atkārto noteiktas darbības izpildi līdz brīdim, kad pārstāj izpildīties noteikts nosacījums.

```python
# For cikla piemērs
a = int(input("Ievadi skaitli: "))
s=0
for i in range(1,a+1):
    s += i 
print(s)
```

```python
#While cikla piemērs
s = 0
a = int(input("Ievadi skaitli: ")) 

while a>0:
    s += a
    a = int(input("Ievadi skaitli: ")) 
    
print("Skaitļu summa:", s)
```

```python
#Break piemērs
var = 10                    
while var > 0:              
    print ("Pašreizējais skaitlis:", var)
    var = var -1
    if (var == 5):
        break
print ("Uz redzēšanos!")
```

```python
#Continue piemērs
var = 10                    
while var > 0:
    var = var -1
    if (var == 5):
        continue
    print ("Pašreizējais skaitlis:", var)
print ("Uz redzēšanos!")
```

## Tekstveida spēles izveide

```python
# nepieciešamo bibliotēku importēšana
import time # atbild par pauzēm

# atbilžu variantu definēšana
atbilde_A = ["A","a"]
atbilde_B = ["B","b"]
atbilde_C = ["C","c"]

nepareizi = ("Atbilde var but tikai A, B vai C") #nepareiza atbilde

#Ievads
def ievads():
    print("Ir vēls vakars un Tu ej pa ielu. Pēkšņi no gaiteņa izskrien zaglis! Ko tu darīsi?")
    time.sleep(1)
   
    # Izvade
    print("""
             A)Mest ar akmeni
             B)Nogulties un cerēt uz labāko
             C)Skriet prom
          """)     
        
    izvele = input(">>> ") # Izvēles izvade
    
    if izvele in atbilde_A:
        print("Tu met ar akmeni!")
        mest_akmeni()
        
    elif izvele in atbilde_B:
        print("Tu nogulies un ceri, ka zaglis tevi nepamanīs")
        print("Zaglis tevi ieradzīja! GAME OVER!")
        
    elif izvele in atbilde_C:
        skriet_prom()
        
    else: #kļūdas apstrāde, ja ievade NAV A,B vai C
        print(nepareizi)
        ievads()
    
def mest_akmeni():
    print("Tu met ar akmeni un trāpi! Bet zaglis turpina tev skriet virsū! Ko tu darīsi tālāk?")
    time.sleep(1)
    print(""" 
              A)Mest vēlvienu akmeni!
              B)Skriet uz mājām!
              """)
    izvele = input(">>> ")
    if izvele in atbilde_A:
        print("Tu met atkal ar akmeni!")
        print("Tu netāpi! GAME OVER!")
    elif izvele in atbilde_B:
        skriet_prom()
    else:
        print(nepareizi)
        mest_akmeni()
        
def skriet_prom():
    print("Tu skrien prom, bet zaglis tev skrien pakaļ! Ko tu tagad darīsi?")
    time.sleep(1)
    print(""" 
              A)Slēpties aiz sienas!
              B)Cīnīties ar zagli!
              """)
    izvele = input(">>> ")
    if izvele in atbilde_A:
        print("Tu slēpies aiz sienas un zaglis tev aiziet garām! Tu izglābies!")
    elif izvele in atbilde_B:
        print("Zaglis tevi nodur! GAME OVER!")
    else:
        print(nepareizi)
        skriet_prom()
ievads()
```
